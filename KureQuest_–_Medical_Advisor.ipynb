```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
import gradio as gr

# Load dataset
df = pd.read_csv('/content/medical_symptoms_dataset.csv')

# Drop any missing values
df = df.dropna()

# Split features and target
X = df['Symptoms']
y = df['Disease']

# Build pipeline
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])

# Train model
model.fit(X, y)
```

```python
def predict_diagnosis(symptom_text):
    prediction = model.predict([symptom_text])[0]
    return f"KureQuest Suggestion: {prediction}"
```

```python
import gradio as gr

ui = gr.Interface(
    fn=predict_diagnosis,
    inputs=gr.Textbox(
        label="üìù Describe Your Symptoms",
        placeholder="e.g. I have sore throat, chills, and fever...",
        lines=4
    ),
    outputs=gr.Textbox(label="üßæ Diagnosis Result"),
    title="üíö KureQuest ‚Äì Medical Advisor",
    description="""
üîé Type your symptoms in plain language. KureQuest will detect known symptoms and provide a likely diagnosis.
""",
    theme="soft",  # ‚úÖ Theme switchable from Gradio UI
    css="""
    body {
        background: linear-gradient(to bottom right, #e6fffa, #ffffff);
        font-family: 'Segoe UI', sans-serif;
    }
    .gradio-container {
        width: 100% !important;
        max-width: 100% !important;
        padding: 2rem;
    }
    h1 {
        color: #065f46;
        font-weight: bold;
        text-align: center;
        font-size: 2.4rem;
    }
    h2 {
        text-align: center;
        color: #047857;
        margin-bottom: 1rem;
        font-size: 1.2rem;
    }
    textarea {
        color: #000 !important;  /* üñäÔ∏è Black text */
        background-color: #f0fdf4;
        border: 2px solid #10b981 !important;
        border-radius: 10px !important;
        padding: 14px;
        font-size: 16px;
    }
    .gr-button {
        background-color: #10b981 !important;  /* ‚úÖ Green button */
        color: white !important;
        font-weight: bold;
        font-size: 16px;
        border-radius: 10px;
        padding: 10px 20px;
    }
    .gr-button:hover {
        background-color: #059669 !important;
    }
    .output-class {
        background-color: #ecfdf5;
        border-left: 4px solid #065f46;
        padding: 1rem;
        font-size: 16px;
        border-radius: 10px;
    }
    """
)

ui.launch()
```
